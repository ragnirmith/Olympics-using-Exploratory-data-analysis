```python
# Maximum number of participants from top 10 countries
top_10_countries=df_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# above table to plot
plt.figure(figsize=(12,6))
plt.title('Overall participation by country',fontweight='bold')
sns.barplot(x=top_10_countries.index,y=top_10_countries);
plt.xlabel('Team')
plt.ylabel('counts')
```

```python
# Age distribution of participants
plt.figure(figsize=(15,6))
plt.title('Age distribution of Atheletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(df_df.Age,bins=np.arange(10,80,2),color='blue',edgecolor='white')
plt.show()
```

```python
# female participation in summer olympics
womenolympics=df_df [(df_df.Sex=='F') & (df_df.Season=='Summer')]
```

```python
womenolympics
```

```python
# countplot for above
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=womenolympics,palette='Spectral')
plt.title("plot of Female Atheletes over time")
```

```python
# lineplot for above
part=womenolympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('plot of female Atheletes over time')
```

```python
# data of gold medal atheletes
goldmedals=df_df.query('Medal=="Gold"')
goldmedals
```

```python
# Age nulls should be removed
goldmedalss=goldmedals[np.isfinite(goldmedals['Age'])]
goldmedalss
```

```python
goldmedal_above_sixty=goldmedals['ID'][(goldmedals['Age'])>60].count()
goldmedal_above_sixty
```

```python
goldmedal_above_sixty_sport=goldmedals['Sport'] [goldmedals['Age']>60]
goldmedal_above_sixty_sport
```

```python
sns.countplot(goldmedal_above_sixty_sport)
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Sport',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title("Gold medals won by people above the age of sixty",fontsize=15,fontweight="bold")
```

```python
GM=goldmedals.Region.value_counts().reset_index(name="Medal").head(5)
GM
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.barplot(x='index',y='Medal',data=GM)
plt.title("plot of Female Atheletes over time")
```

```python
# plot for above data
totalgoldmedals=goldmedals.Region.value_counts().reset_index(name="Medal").head(6)
g=sns.catplot(x='index',y='Medal',data=totalgoldmedals,height=5,kind='bar',palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 countries')
g.set_ylabels('Number of Medals')
plt.title('Goldmedals per country')
```
